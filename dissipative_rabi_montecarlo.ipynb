## 1. Rabi oscillation

Rabi oscillation occurs in two-level systems under external periodic drives. One of the classic examples is the **atomic light-matter interaction**, where an atom is driven by electromagnetic field.

### 1.1. Model of two-level system and light-matter interaction

We first determine the model for our study. The atomic states are modeled by a two-level system
$$H_0 = -\frac{\omega_0}{2}\begin{pmatrix}1&0\\0&-1\end{pmatrix},$$
where $\omega_0$ is the energy difference between the ground and excited states. The eigenstates of this system are $\ket{0}$ and $\ket{1}$ with energies $E_{0,1} = \mp\omega_0/2$. Meanwhile, the electromagnetic field is modeled by an off-diagonal coupling
$$H_1(t) = \Omega\cos\omega t\begin{pmatrix}0&1\\1&0\end{pmatrix}.$$
Here $\Omega$ is the **Rabi frequency** that characterizes the amplitude of the electromagnetic field, and $\omega$ is the oscillating frequency. For simplicity, we have assumed the Rabi frequency as real $\Omega\in\mathbb R$. The total Hamiltonian contains both terms
$$H(t) = H_0+H_1(t).$$

The Schrödinger equation of this system takes the form
$$i\frac{d}{dt}\ket{\psi(t)} = H(t)\ket{\psi(t)},$$
where we have assumed $\hbar = 1$. Expressing the wavefunction in the eigenbasis of $H_0$
$$\ket{\psi(t)} = \begin{pmatrix}a_0(t)e^{-iE_0t}\\a_1(t)e^{-iE_1t}\end{pmatrix},$$
we can derive an effective Schrödinger equation
$$i\frac{d}{dt}\ket{\psi'(t)} = H'(t)\ket{\psi'(t)}$$
with the effective state and Hamiltonian
$$\ket{\psi'(t)} = \begin{pmatrix}a_0(t)\\a_1(t)\end{pmatrix},\quad
H'(t) = \begin{pmatrix}0&\Omega e^{-i\omega_0t}\cos\omega t\\\Omega e^{i\omega_0t}\cos\omega t&0\end{pmatrix}.$$
Note that the contribution from the static Hamiltonian $H_0$ is cancelled by the energy terms from the time derivatives. The effective Hamiltonian can be rewritten as
$$H'(t) = \frac{1}{2}\Omega \begin{pmatrix}0&e^{i(\omega-\omega_0)t}+e^{-i(\omega+\omega_0)t}\\e^{i(\omega+\omega_0)t}+e^{-i(\omega-\omega_0)t}&0\end{pmatrix}.$$
We can adopt the *rotating wave approximation (RWA)* by neglecting the fast oscillating terms with frequency $\omega+\omega_0$. The result is
$$H'(t) = \frac{1}{2}\Omega \begin{pmatrix}0&e^{i\Delta t}\\e^{-i\Delta t}&0\end{pmatrix},$$
where the difference of electromagnetic field frequency and energy gap $\Delta = \omega-\omega_0$ is known as the **detuning**.

We further transform the Schrödinger equation to a time-dependent basis
$$\ket{\psi'(t)} = U(t)\ket{\tilde\psi(t)}, \quad U(t) = \begin{pmatrix}1&0\\0&e^{-i\Delta t}\end{pmatrix}.$$
This transformation leads to
$$i\frac{d}{dt}\ket{\tilde\psi(t)} = \tilde H\ket{\tilde\psi(t)},\tag{1}$$
where the Hamiltonian becomes independent of time
$$\tilde H = \begin{pmatrix}0&\Omega/2\\\Omega/2&-\Delta\end{pmatrix}.\tag{2}$$
We will use this time-independent Hamiltonian to for our later study of Rabi oscillation.

### 1.2. Rabi oscillation: Analytic approach

We have obtained a time-independent form of the Hamiltonian for the driven two-level system. Now we can study the time evolution of the state
$$\ket{\tilde\psi(t)} = \begin{pmatrix}\tilde a_0(t)\\\tilde a_1(t)\end{pmatrix}.$$
Here we take the ground state with $(\tilde a_0(0), \tilde a_1(0)) = (1, 0)$ as the initial condition.

To solve the time evolution, we first obtain the eigenstates of the Hamiltonian $\tilde H$. The eigenvalues can be calculated directly as
$$\tilde E_{0,1} = \frac{1}{2}\left(-\Delta\pm\Omega_R\right),$$
where the definition of a **generalized Rabi frequency**
$$\Omega_R = \sqrt{\Delta^2+\Omega^2}$$
will become clear later. The eigenstates take the form
$$\ket{\tilde E_{0,1}} = b_{0,1}\begin{pmatrix}\Omega\\-\Delta\pm\Omega_R\end{pmatrix},\quad b_{0,1} = \frac{1}{\sqrt{2\Omega_R\mp2\Delta\Omega_R}}.$$

Now we can use the eigenstates to express the general state
$$\ket{\tilde\psi(t)} = c_0\ket{E_0}e^{-iE_0t}+c_1\ket{E_1}e^{-iE_1t},\quad |c_0^2|+|c_1|^2=1.$$
The initial ground state determines the relation
$$\left\{\begin{aligned}&c_0b_0+c_1b_1 = \frac{1}{\Omega},\\&c_0b_0-c_1b_1 = \frac{\Delta}{\Omega\Omega_R}.\end{aligned}\right.$$
The coefficients can then be derived as
$$c_{0,1} = \frac{1}{2b_{0,1}\Omega}\left(1\pm\frac{\Delta}{\Omega_R}\right).$$
Expanding the time-dependent state with these results
$$\ket{\tilde\psi(t)} = \frac{1}{2\Omega}e^{i\Delta t/2}\left[\left(1+\frac{\Delta}{\Omega_R}\right)\begin{pmatrix}\Omega\\-\Delta+\Omega_R\end{pmatrix}e^{-i\Omega_Rt/2}+\left(1-\frac{\Delta}{\Omega_R}\right)\begin{pmatrix}\Omega\\-\Delta-\Omega_R\end{pmatrix}e^{i\Omega_Rt/2}\right],$$
we obtain the desired solution of the time-evolving state
$$\ket{\tilde\psi(t)} = e^{i\Delta t/2}\left[\cos\frac{\Omega_R}{2}t\begin{pmatrix}1\\0\end{pmatrix}-i\frac{1}{\Omega_R}\sin\frac{\Omega_R}{2}t\begin{pmatrix}\Delta\\\Omega\end{pmatrix}\right].$$

It is now clear that the state undergoes an oscillation between the ground and excited states. Notably, the oscillation frequency of populations, such as the one for the excited state
$$|\tilde a_1(t)|^2 = \frac{\Omega^2}{\Omega_R^2}\sin^2\frac{\Omega_R}{2}t = \frac{\Omega^2}{2\Omega_R^2}(1-\cos\Omega_Rt),$$
is determined by the generalized Rabi frequency $\Omega_R$. The detuning plays an essential role in the oscillation behavior of the state. When the detuning vanishes $\Delta = 0$, that is, the electromagnetic-field frequency matches the energy gap $\omega = \omega_0$, the state can populate completely in the excited state at $t = \pi/\Omega_R$. This effect is known as the **resonance** between the two-level system and the electromagnetic field. At this resonance, the generalized Rabi frequency is equal to the Rabi frequency $\Omega_R = \Omega$. When the electromagnetic field is detuned away, the oscillation always leaves residual populations in the ground state.

Let us plot the population of the excited state. Here we choose the energy gap $\omega_0 = 1$, Rabi frequency $\Omega = 0.5$, and leave the detuning as a tunable parameter:

```python
import math

# Set the energy gap.
omega_0 = 1
# Set the Rabi frequency.
Omega = 0.5
# Define the function for the excited-state population.
def excited_state_population(t, omega_0, Omega, Delta):
    Omega_R = math.sqrt(Delta**2 + Omega**2)
    return (Omega**2 / (2 * Omega_R**2)) * (1 - math.cos(Omega_R * t))
```

The results are plotted as follows:

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a list of time steps.
t_i, t_f = 0, 20
ts = np.linspace(t_i, t_f, 1000)
# Plot the excited-state populations at different detunings.
Deltas = [-1., -0.5, 0., 0.5, 1.]
for Delta in Deltas:
    plt.plot(ts, [excited_state_population(t, omega_0, Omega, Delta) for t in ts], label = r"$\Delta =$"+f"{Delta}")

# Finish the plot.
plt.title("Rabi oscillation: Analytic result")
plt.xlim(t_i, t_f)
plt.ylim(0, 1)
plt.xlabel(r"$t$")
plt.ylabel(r"$|\tilde a_1(t)|^2$")
plt.legend()
plt.show()
```

We can notice that the oscillation is indeed maximal at the resonance $\Delta = 0$, while the positive and negative detunings give identical results at the same absolute values.

### 1.3. Rabi oscillation: Numerical approach

We have analytically derived the Rabi oscillation in a two-level system under electromagnetic field. We can also obtain the result numerically, which will serve as an important starting point when we include spontaneous emission. To compute the Rabi oscillation numerically, we take the Schrödinger equation (1) with time-independent Hamiltonian (2)
$$\frac{d}{dt}\ket{\tilde\psi(t)} = f(\ket{\tilde\psi(t)}),\quad f(\ket{\tilde\psi(t)}) = -i\tilde H\ket{\tilde\psi(t)}.$$
With this ordinary differential equation (ODE), we can use the ``solve_ivp`` function in ``scipy.integrate`` to solve the time evolution.

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt

# Set the energy gap.
omega_0 = 1
# Set the Rabi frequency.
Omega = 0.5
# Set the detunings.
Deltas = [-1., -0.5, 0., 0.5, 1.]
# Set the initial condition as the ground state.
psi_0 = np.array([1., 0.], dtype = complex)
# Set the time span including initial and final times, and the time points for evaluations.
t_i, t_f = 0, 20
t_span = [t_i, t_f]
t_eval = np.linspace(t_i, t_f, 1000)

# Compute the time evolution at various detunings.
for Delta in Deltas:
    # Set the Hamiltonian.
    H = np.array([[0, Omega/2], [Omega/2, -Delta]], dtype = complex)
    # Define the function of the differential equation.
    def fun(t, psi):
        return -1.j * np.dot(H, psi)
    sol = solve_ivp(fun, t_span, psi_0, t_eval = t_eval)
    plt.plot(t_eval, np.pow(abs(sol.y[1]), 2), label = r"$\Delta =$"+f"{Delta}")

# Finish the plot.
plt.title("Rabi oscillation: Numerical result")
plt.xlim(t_i, t_f)
plt.ylim(0, 1)
plt.xlabel(r"$t$")
plt.ylabel(r"$|\tilde a_1(t)|^2$")
plt.legend()
plt.show()
```

We get an identical result to our previous analytic approach.

## 2. Dissipative Rabi oscillation: Spontaneous emission

So far, we have considered the Rabi oscillation in a two-level system under electromagnetic field. This process captures the **stimulated absoprtion and emission**. The light-matter interaction involves only one type of photons, whose wave vector and polarization are determined by the electromagnetic field. However, there exists another type of light-matter interaction, known as the **spontaneous emission**. Here the two-level system couples to the zero-point vacuum state $\ket{0_\text{ph}}$ of the photons, such that the excited state goes back to the gound state by spontaneously emitting a photon of arbitrary wave vector and polarization.

### 2.1. Non-Hermitian Hamiltonian for dissipative spontaneous emission

We may understand the process in the framework of discrete time evolution [2]. At each infinitesimal time step $t\rightarrow t+dt$, a photon may be emitted spontaneously
$$\ket{\tilde\psi(t)}\otimes\ket{0_\text{ph}} \rightarrow \ket{\tilde\psi^0(t+dt)}\otimes\ket{0_\text{ph}} + \alpha_1\ket{0}\otimes\ket{1_\text{ph}}.\tag{3}$$
The first term represents the time evolution without spontaneous emission, where $\ket{\tilde\psi^0(t+dt)}$ is the corresponding time-evolved two-level state. Meanwhile, the second term captures the dissipation where the excited state $\ket{1}$ spontaneously emits a photon to the vacuum $\ket{1_\text{ph}}$. The probability of spontaneous emission is proportional to the population of the excited state
$$dp = |\alpha_1|^2 = dt\Gamma|\tilde a_1(t)|^2,$$
where $\Gamma$ is the decay rate. Correspondingly, the probability without spontaneous emission 
$$\braket{\tilde\psi^0(t+dt)|\tilde\psi^0(t+dt)} = 1-dp = 1-dt\Gamma|\tilde a_1(t)|^2$$
reduces with time.

We are interested in a description of the dissipative two-level system, without needing the knowledge of the photon vacuum state. This may be achieved by introducing a non-Hermitian Hamiltonian
$$\tilde H' = \begin{pmatrix}0&\Omega/2\\\Omega/2&-\Delta-i\Gamma/2\end{pmatrix}$$
to describe the time evolution $\ket{\tilde\psi(t)}\rightarrow\ket{\tilde\psi^0(t+dt)}$. The Hamiltonian gives a reduced wavefunction norm under time evolution
$$\braket{\tilde\psi^0(t+dt)|\tilde\psi^0(t+dt)}
= \braket{\tilde\psi(t)|\{1+i\tilde Hdt+i[(\tilde H')^\dagger-\tilde H]dt\}\{1-i\tilde Hdt-i[\tilde H'-\tilde H]dt\}|\tilde\psi(t)}
\approx 1-dt\Gamma|\tilde a_1(t)|^2,$$
consistent with the probability without spontaneous emission. Therefore, we can start from this model to study the dissipation effects from spontaneous emission.

### 2.2. Monte Carlo wavefunction approach

An interesting approach to solving this problem is the utilization of Monte Carlo framework. In each discrete time step (3), we can assume that the photon is either spontaneously emitted or not. The choice is determined by a random number $p\in[0,1]$. If $p>dp$, we assume that the state evolves without spontaneous emission. In this case, we renormalize the state
$$\ket{\tilde\psi^0(t+dt)} \rightarrow \ket{\tilde\psi(t+dt)} = \frac{\ket{\tilde\psi^0(t+dt)}}{\braket{\tilde\psi^0(t+dt)|\tilde\psi^0(t+dt)}}$$
and move on to the next time step. However, if $p<dp$, we reset the state to the ground state
$$\ket{\tilde\psi^0(t+dt)} \rightarrow \ket{\tilde\psi(t+dt)} = \ket{0}$$
and continue the time evolution from there. This kind of abrupt reset is called a **quantum jump**. The involvement of this randomness in time evolution allows a way to tackle the spontaneous nature of the emission process. Note that the inclusion of randomness means that we need to do a randomness average. Each single-shot time evolution represents an individual **quantum trajectory** of a single atom, and the randomness average describes the behavior of an ensemble of atoms [1].

Let us now examine the result of this Monte Carlo wavefunction approach. Here we study the resonant case with $\Delta = 0$, where the Rabi oscillation can bring the ground state completely to the excited state. The parameters we use in our computation are as follows:

```python
import numpy as np

# Set the energy gap.
omega_0 = 1
# Set the Rabi frequency.
Omega = 0.5
# Set the decay rate.
Gamma = 0.1
# Set the detunings.
Delta = 0.
# Set the initial condition as the ground state.
psi_0 = np.array([1., 0.], dtype = complex)
# Set the time span including initial and final times, and the time points for evaluations.
t_i, t_f = 0, 100
t_eval = np.linspace(t_i, t_f, 10000)
dt = t_eval[1] - t_eval[0]
```

Since the trajectory can now experience discontinuities conditionally, we need to write our own code to compute the integration appropriately. Here we will use the fourth-order Runge-Kutta method (RK4) [3]. Consider a differential equation of a function $y(t)$
$$\frac{dy}{dt} = f(t,y)$$
with an initial condition $y(t=0)=y_0$. For each time step $t_m\rightarrow t_{m+1} = t_m+\Delta t$, the RK4 method moves $y_m$ to $y_{m+1}$ as follows:
1. Compute $k_1 = f(t_m, y_m)$.
2. Compute $k_2 = f(t_m+\Delta t/2, y_m+[\Delta t/2]k_1)$.
3. Compute $k_3 = f(t_m+\Delta t/2, y_m+[\Delta t/2]k_2)$.
4. Compute $k_4 = f(t_m+\Delta t, y_m+\Delta tk_3)$.
5. Compute $y_{m+1} = y_m+(\Delta t/6)(k_1+2k_2+2k_3+k_4)$.

```python
import random
import math

def time_evol(dt, psi_m, H):
    # Implement the RK4 method.
    k1 = -1.j * np.dot(H, psi_m)
    k2 = -1.j * np.dot(H, psi_m + (dt/2) * k1)
    k3 = -1.j * np.dot(H, psi_m + (dt/2) * k2)
    k4 = -1.j * np.dot(H, psi_m + dt * k3)
    psi_m1 = psi_m + (dt / 6) * (k1 + 2 * k2 + 2 * k3 + k4)
    # Compute the norm square of the wavefunction psi_m1.
    normsq_psi_m1 = np.linalg.norm(psi_m1)**2
    # Generate a random number between 0 and 1.
    p = random.uniform(0., 1.)
    # Determine whether the spontaneous emission occurs.
    if p > 1 - normsq_psi_m1:
        psi_m1 = (1/math.sqrt(normsq_psi_m1)) * psi_m1
    else:
        psi_m1 = np.array([1., 0.])
    return psi_m1
```

Having defined the time evolution function, we can now compute the quantum trajectory and examine the dissipative effects of spontaneous emission on Rabi oscillation. We first study the case of a single quantum trajectory.

```python
import matplotlib.pyplot as plt

# Set the Hamiltonian.
H = np.array([[0, Omega/2], [Omega/2, -Delta - 1.j * Gamma/2]], dtype = complex)

# Compute the time evolution.
psi_eval = [psi_0]
psi_m = psi_0
for i in range(1, len(t_eval)):
    psi_m1 = time_evol(dt, psi_m, H)
    psi_eval.append(psi_m1)
    psi_m = psi_m1.copy()
psi_eval = np.array(psi_eval).T

# Plot the time evolution result.
plt.plot(t_eval, np.pow(abs(psi_eval[1]), 2))
# Finish the plot.
plt.title("Dissipative Rabi oscillation: Single quantum trajectory")
plt.xlim(t_i, t_f)
plt.ylim(0, 1)
plt.xlabel(r"$t$")
plt.ylabel(r"$|\tilde a_1(t)|^2$")
plt.show()
```

We can see that the quantum trajectory involves multiple quantum jumps, where spontaneous emission occurs and the excited state jumps back to the ground state. This kind of quantum trajectories describe the behavior of a single atom in the electromagnetic field.

Now let us do a randomness average over $1000$ quantum trajectories.

```python
import matplotlib.pyplot as plt

# Set the Hamiltonian.
H = np.array([[0, Omega/2], [Omega/2, -Delta - 1.j * Gamma/2]], dtype = complex)

# Randomness average over 1000 quantum trajectories.
psi_abssq_eval_avg = np.zeros((2, len(t_eval)), dtype = complex)
N_rand = 1000
for n_rand in range(N_rand):
    # Compute the time evolution.
    psi_eval = [psi_0]
    psi_m = psi_0
    for i in range(1, len(t_eval)):
        psi_m1 = time_evol(dt, psi_m, H)
        psi_eval.append(psi_m1)
        psi_m = psi_m1.copy()
    psi_eval = np.array(psi_eval).T
    psi_abssq_eval_avg += np.pow(np.abs(psi_eval), 2)
psi_abssq_eval_avg /= N_rand

# Plot the time evolution result.
plt.plot(t_eval, np.pow(psi_abssq_eval_avg[1], 2))
# Finish the plot.
plt.title("Dissipative Rabi oscillation: Averaged quantum trajectory")
plt.xlim(t_i, t_f)
plt.ylim(0, 1)
plt.xlabel(r"$t$")
plt.ylabel(r"$|\tilde a_1(t)|^2$")
plt.show()
```

Interestingly, the randomness-averaged quantum trajectory shows a strongly damped Rabi oscillation. The damped oscillation evolves to a steady state with a nonzero population in the excited state. This kind of quantum trajectory describes the overall behavior of an ensemble of atoms under electromagnetic field.

## 3. Conclusion

In this Notebook, we have studied the Rabi oscillation in a two-level system under periodic drive. This model captures the light-matter interaction in the atoms under electromagnetic field. We have shown that the Rabi oscillation can bring the ground state to the excited state periodically, where the maximal population is determined by the detuning. We have further investigated the effect of spontaneous emission, which leads to dissipations that draws the excited state back to the ground state. The spontaneous nature of this process is analyzed by a Monte Carlo wavefunction approach, where quantum jumps in single-shot quantum trajectories lead to strong damping under ensemble average. This approach provides a transparent interpretation of how light-matter interaction affects individual atoms and their ensembles.
